```python
### WR64 nexus cell.section::: cell.sections[72].recVList[0] 
### WR71 nexus cell.section::: cell.sections[24].recVList[0]

def run_gillis_protocol_v1(row):
    import neuron
    h = neuron.h
    output = {}
    stim_amp = 0.1
    no_spike = True
    stim_dur = 500 
    
    # get simulator and evaluator object for a specific morphology
    morphology = row['morphology_id']
    
    s = get_simulator(morphology)
    e = get_evaluator(morphology)
    p = row[params_py3]
    
    while no_spike == True:

        with I.silence_stdout:
            neup = s.setup.get_cell_params_with_default_sim_prams(p)
            cell = I.scp.create_cell(neup.neuron)
            nex_sec = get_index_of_bifurcation_seciton(cell)
        iclamp = h.IClamp(0.5, sec=cell.sections[nex_sec])
        iclamp.delay = 295 
        iclamp.dur = stim_dur 
        iclamp.amp = stim_amp
        
        sim_params = I.scp.NTParameterSet({'tStart': 0.0,'tStop': 295+stim_dur+10,'dt': 0.025,'Vinit': -75.0,'T': 34.0,'recordingSites': []})
        I.scp.init_neuron_run(sim_params, vardt = True)

        t = I.np.array(cell.tVec) # time
        v_soma = I.np.array(cell.sections[0].recVList[0]) # soma voltage trace
        v_nexus = I.np.array(cell.sections[nex_sec].recVList[0]) # nexus voltage trace
        
        if max(v_soma)<0:
            stim_amp += 0.1
        else:
            print(stim_amp)
            no_spike = False
    
    output['stim_amp'] = stim_amp
  
    # get simulator and evaluator object for a specific morphology
    with I.silence_stdout:
        neup = s.setup.get_cell_params_with_default_sim_prams(p)
        cell = I.scp.create_cell(neup.neuron)
    
    iclamp = h.IClamp(0.5, sec=cell.sections[nex_sec])
    iclamp.delay = 295 
    iclamp.dur = stim_dur 
    iclamp.amp = stim_amp 
    
    sim_params = I.scp.NTParameterSet({'tStart': 0.0,'tStop': 295+stim_dur+10,'dt': 0.025,'Vinit': -75.0,'T': 34.0,'recordingSites': []})
    I.scp.init_neuron_run(sim_params, vardt = True)
    
    # control
    t = I.np.array(cell.tVec) # time
    v_soma = I.np.array(cell.sections[0].recVList[0]) # soma voltage trace
    v_nexus = I.np.array(cell.sections[nex_sec].recVList[0]) # nexus voltage trace
    # no stimulus set up, steady state reached at around 200 ms
    
    output['t_control'] = t
    output['v_soma_control'] = v_soma
    output['v_nexus_control'] = v_nexus
    
    ##################################
    ############ Modulation ##########
    ##################################
    
    HVA_multiplyer = 1.2
    SKv3_1_multiplyer = 1
    
    neup = s.setup.get_cell_params_with_default_sim_prams(p)
    for seg in[neup.neuron.AIS,neup.neuron.ApicalDendrite,neup.neuron.Soma]:
        seg['mechanisms']['range']['Ca_HVA']['gCa_HVAbar'] *= HVA_multiplyer
    # for seg in[neup.neuron.ApicalDendrite]:
    #     seg['mechanisms']['range']['SKv3_1']['gSKv3_1bar'] *= SKv3_1_multiplyer
    with I.silence_stdout:
        cell = I.scp.create_cell(neup.neuron)
        iclamp = h.IClamp(0.5, sec=cell.sections[nex_sec])
        iclamp.delay = 295 
        iclamp.dur = stim_dur 
        iclamp.amp = stim_amp 
        sim_params = I.scp.NTParameterSet({'tStart': 0.0,'tStop': 295+stim_dur+10,'dt': 0.025,'Vinit': -75.0,'T': 34.0,'recordingSites': []})
    I.scp.init_neuron_run(sim_params, vardt = True)
    
    t = I.np.array(cell.tVec) # time
    v_soma = I.np.array(cell.sections[0].recVList[0]) # soma voltage trace
    v_nexus = I.np.array(cell.sections[nex_sec].recVList[0]) # nexus voltage trace
    
    output['t_modulation'] = t
    output['v_soma_modulation'] = v_soma
    output['v_nexus_modulationl'] = v_nexus
    
    return output
```

```python
@I.dask.delayed
def run_chunk_and_save(chunk, outdir, chunk_id):
    output = {}
    for name, row in chunk.iterrows():
        output[name] = run_gillis_protocol_v1(row)
    outpath = I.os.path.join(outdir, '{}.pickle'.format(chunk_id))
    with open(outpath, 'wb') as f:
        I.cloudpickle.dump(output, f)
```
